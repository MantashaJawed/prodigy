```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('cats_dataset.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=["Gender"], inplace=True)
data["Color"].fillna("unknowm", inplace=True)
data["Age (Years)"].fillna(data["Age (Years)"].mean(), inplace=True)

```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python

plt.figure(figsize=(6, 3))
sns.histplot(data["Age (Years)"], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age (Years)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=data, x="Breed", hue="Gender")
plt.title("cats")
plt.xlabel("Breed")
plt.ylabel("Count")
plt.legend(title="Gender", loc="upper right")
plt.show()
```

```python


plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x="Age (Years)", y="Breed", hue="Gender")
plt.title("Scatter Plot of Breed and their age")
plt.xlabel("Age")
plt.ylabel("Breed Name")

plt.show()
```
